```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
df= pd.read_csv('student_performance.csv')
df
```

```python
df.info()
```

```python
df['final_score']=(df['math_score']+df['writing_score']+df['reading_score'])/3
```

```python
# creating a new column that checks if the student has passed 0r failed
lamb= lambda final_score: 'Passed' if final_score>50 else 'Failed'
df['status']=df['final_score'].apply(lamb)
df
```

```python
df
```

```python
# creatinting a new column grades to obtain grades of students from their final_score
def check_grade(final_score):
    if final_score>=80:
        return 'A'
    elif final_score>=60:
        return 'B'
    elif final_score>=50:
        return 'C' 
    elif final_score>=40:
        return 'D'
    else:
        return 'F'  
df['grade']=df['final_score']   .apply(check_grade)   
df  
```

```python
from scipy import stats 
x=df['math_score']
y=df['reading_score']
slope, intercept, r, p, std_err = stats.linregress(x, y)
def myfunction(x):
    return slope * x + intercept
mymodel = list(map(myfunction, x))
plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show
```
